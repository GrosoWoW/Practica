## Valorización completa

```python
infomio = derivado10.flujos_valorizados
print("Fondo DMLVESTRAT")
print("Activo")
act = sum(infomio[infomio["ActivoPasivo"] == 1].ValorPresenteMonFlujo)
print("Valor activo:", act)
print()
print("Pasivo")
pas = sum(infomio[infomio["ActivoPasivo"] == -1].ValorPresenteMonFlujo)
print("Valor pasivo:", pas)
valor10 = act-pas
print()
print("Total")
print(valor10)
```

# Mensaje para el correo

```python
mensaje =  '''Estimados,

Estas son las valorizaciones al @dia de @mes.

    Fondo FMLVESTRAT recibe tasa fija 1.45% a 2 años por $10.000.000.000

    @valor10 CLP  

    Fondo CFILVDE recibe tasa fija 1.45% a 2 años por $5.000.000.000

    @valor5 CLP 


Saludos
'''

meses = {1:"Enero",
        2:"Febrero",
        3: "Marzo",
        4: "Abril",
        5: "Mayo",
        6: "Junio",
        7: "Julio",
        8: "Agosto",
        9: "Septiembre",
        10: "Octubre",
        11: "Noviembre",
        12: "Diciembre"}



mensaje = mensaje.replace("@dia", str(fecha.day))

# El mes se pone en espanhol
mensaje = mensaje.replace("@mes", meses[fecha.month] )

mensaje = mensaje.replace("@valor10", str(round(valor10,2)))

mensaje = mensaje.replace("@valor5", str(round(valor5,2)))


print(mensaje)
```
